# DUMMY VARIABLES

```python
import pandas as pd
df = pd.read_csv('homeprices1.csv')
df
```

GET DUMMIES METHOD


```python
dummies = pd.get_dummies(df.town)
dummies
```

CONCATING THIS DF TO ORIGINAL DF

```python
merged = pd.concat([df,dummies],axis=1)
merged
```

```python
merged.drop('town',inplace=True,axis=1)
merged
```

DROPPING ONE OF  THE COLUMN TO ENSURE MULTI-COLLINEARITY DOESNT AFFECT

```python
final = merged.drop(['west windsor'],axis=1)
final
```

CREATING LINEAR REGRESSION OBJECT

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

NOW DECLARING X AND Y FOR TRAINING.

```python
X = final.drop(['price'],axis = 1)
X
```

```python
Y = final.price
Y
```

```python
model.fit(X,Y)
```

PREDICTING THE PRICE

```python
model.predict([[2800,0,1]])
# 0,1 INDICATES THE PREICE IN ROBSVILLE AND 1,0 IN monroe township
```

```python
model.predict([[2800,0,0]])
# IN WEST WINSOR
```

```python
model.predict([[2400,0,0]])
# IN MINSORE
```

CALCULATING THE ACCURACY

```python
model.score(X,Y)
```
